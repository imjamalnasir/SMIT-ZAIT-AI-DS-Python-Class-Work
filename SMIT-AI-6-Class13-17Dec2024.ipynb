```python
class Dog():
    def __init__(self,name,age,breed,bite):
        self.name = name
        self.age = age
        self.breed = breed
        self.bite = bite
    def bark(self):
        print(f"{self.name} is barking")
    def describeDog(self):
        print(f"""
          Dog Name: {self.name}
          Dog Breed: {self.breed}
          Dog Age: {self.age}
          Dog Bite: {self.bite}
        """)
```

```python
class Car():
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0                
        
    def get_descriptive_name(self):
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()
        
    def read_odometer(self):
        print("This car has " + str(self.odometer_reading) + " miles on it.")
        
    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")
            
    def increment_odometer(self, miles):
        self.odometer_reading += miles
```

```python
car1 = Car('Honda', '2022', '2024')
```

```python
car1.get_descriptive_name()
```

```python
car1.read_odometer()
```

```python
car1.increment_odometer(123)
```

```python
car1.read_odometer()
```

```python
car1.increment_odometer(10)
```

```python
car1.read_odometer()
```

```python
car1.update_odometer(500)
```
